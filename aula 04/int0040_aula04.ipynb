### Python - Comando eval
O comando eval pode ser utilizado para a construção de comandos customizados para serem executados pelo interpretador python. Segue um exemplo de uma aplicação de respostas automatizadas montadas a partir de um conjunto de respostas pré-determinadas. No entanto, poderiam ser gerados outros comandos mais complexos.

**Atenção** - Esse tipo de abordagem pode ser perigoso a depender de como as entradas são processadas, podendo gerar ameaças caso não haja um processamento das entradas antes da execução. 

> Um exemplo de vulnerabilidade "similar" acontece no caso da montagem de consultas (queries) [SQL](https://en.wikipedia.org/wiki/SQL) em aplicações web conhecida como [SQL injection](https://portswigger.net/web-security/sql-injection).

```python
#chatbot cearense
import random

s = input('Escreva um nome:')
respostas = ['Deixa de ser besta,','Ai dentu','Bora botar um buneco grande,']

i = random.randint(0,len(respostas)-1)
eval('print("{} {}")'.format(respostas[i],s))
```

- Passo 2: Conversão do xml em dict e arquivamento como JSON

```python
import xmltodict
import json
import os

print(os.getcwd())


with open("/content/drive/MyDrive/tecnicas_computacionais/nota.xml") as f:
    text = f.read()

print(text)
dictionary = xmltodict.parse(text)

json_object = json.dumps(dictionary)
print(json_object)

with open("/content/drive/MyDrive/tecnicas_computacionais/nota.json","w") as f:
    f.write(json_object)

for k,v in dictionary.items():
    print(k,v)
```
