- Passo 2: Conversão do xml em dict e arquivamento como JSON

```python
import xmltodict
import json
import os

print(os.getcwd())


with open("/content/drive/MyDrive/tecnicas_computacionais/nota.xml") as f:
    text = f.read()

print(text)
dictionary = xmltodict.parse(text)

json_object = json.dumps(dictionary)
print(json_object)

with open("/content/drive/MyDrive/tecnicas_computacionais/nota.json","w") as f:
    f.write(json_object)

for k,v in dictionary.items():
    print(k,v)

d = dictionary
for k,v in d['nfeProc'].items():
  print('[{}] - {}'.format(k,v))
```
